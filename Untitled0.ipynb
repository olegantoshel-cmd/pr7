```python
import re
data_text = """
Посетите наши страницы:
Email: info@example.com (главный), support-team@corp.net (поддержка)
Дата: 25-05-2023, 01/01/2024, 12.02.2025.
Коды товаров: ID_PROD-1001, ITEM-20-A, ID-300-B.
Суммы: $150.50, 20 EUR, 500 RUB, £75.
"""
date_pattern = r'(\d{2})\.(\d{2})\.(\d{4})'

def convert_date_format(match_obj):
    day = match_obj.group(1)
    month = match_obj.group(2)
    year = match_obj.group(3)
    return f"{year}-{month}-{day}"

modified_text = re.sub(date_pattern, convert_date_format, data_text)

print("Текст с замененными датами:")
print(modified_text)
```

```python
import re
data_text = """
Посетите наши страницы:
Email: info@example.com (главный), support-team@corp.net (поддержка)
Дата: 25-05-2023, 01/01/2024, 12.02.2025.
Коды товаров: ID_PROD-1001, ITEM-20-A, ID-300-B.
Суммы: $150.50, 20 EUR, 500 RUB, £75.
"""
currency_pattern = r'\d+\s*(?:EUR|RUB)'
eur_rub_amounts = re.findall(currency_pattern, data_text)
print("Суммы в EUR или RUB:", eur_rub_amounts)
```
